# Moroney Course Exercise One
Create a neural net (just one node) to predict home value from number of bedrooms if:

$ price = 50K + bedrooms * 50K $

Only need a single node since nodes themselves can do linear operations

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras 
```

```python
# creates network with single node, with input as a single value (1x1 array)
model = tf.keras.Sequential(
    keras.layers.Dense(
        units=1, input_shape=[1]
    )
)
```

```python
# use stochastic gradient descent in backprop to optimize loss, and use mse as loss function to minimize 
model.compile(
    optimizer = 'sgd', loss='mean_squared_error'
)
```

```python
br = np.array([0, 1, 2, 3, 4, 5], dtype=float)
price = np.array([50, 100, 150, 200, 250, 300], dtype=float)
```

```python
model.fit(br, price, epochs=500)
```

```python
print(model.predict([7]))
```
